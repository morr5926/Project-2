```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={"Engine HP":"HP", "Engine Cylinders": "Cylinders", "Transmission Type":"Transmission", "Driven_Wheels": "Drive Mode", "highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price"})
df.head(5)
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df = df.dropna() #Dropping the missing values

df.count()
```

```python
# Calculate quantiles on numeric columns only
numeric_df = df.select_dtypes(include=[np.number])
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Filter out the outliers on the original DataFrame
filter_condition = ~((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)
df = df[filter_condition]

print(df.shape)
```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
df_numeric = df.select_dtypes(include=[np.number])
plt.figure(figsize=(10,5))
c=df_numeric.corr()
sns.heatmap(c, cmap="BrBG", annot=True)
c
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```

```python
#Calclulate the year averages for MPG
yearly_average_mpgH = df.groupby('Year')['MPG-H'].mean()
display(yearly_average_mpgH)
yearly_average_mpgC = df.groupby('Year')['MPG-C'].mean()
display(yearly_average_mpgC)

# Plot the yearly averages
plt.plot(yearly_average_mpgH.index, yearly_average_mpgH.values, label='Average MPG-Highway')
plt.plot(yearly_average_mpgC.index, yearly_average_mpgC.values, label='Average MPG-City')
plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.title('Yearly Average MPG Plot')
plt.legend()
```
